# ECT Fundamental Constants — Parameter Space

**Paper §5** | Three condensate parameters → three constants

| Parameter | → | Constant |
|---|---|---|
| α=2β | → | c (speed of light) |
| v₀=M̄_Pl | → | G_N (Newton) |
| v₀, λ | → | ℏ (Planck) |

**Lorentzian window:** $\beta < \alpha < 4\beta$

```python
import numpy as np, matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider
%matplotlib inline

G_ref = 6.674e-11; hbar_ref = 1.0546e-34; c_ref = 2.998e8
GeV_kg = 1.7827e-27

def plot_params(log10_v0=18.39, alpha=2.0, beta=1.0):
    v0_GeV = 10**log10_v0
    v0_J = v0_GeV * GeV_kg * c_ref**2
    ok_lorentz = beta < alpha < 4*beta
    status = '✓ Lorentzian' if ok_lorentz else '✗ NOT Lorentzian'
    
    if alpha > beta:
        c_star = np.sqrt(beta/(alpha-beta))
        G_N = hbar_ref*c_ref**3/(8*np.pi*v0_J**2*(alpha-beta))
        G_ratio = G_N/G_ref
    else:
        c_star = float('nan'); G_ratio = float('nan')
    
    fig,axes = plt.subplots(1,3,figsize=(14,4))
    
    # c* vs alpha
    if beta > 0:
        a_range = np.linspace(beta*1.01, beta*3.9, 300)
        c_range = np.sqrt(beta/(a_range-beta))
        axes[0].semilogy(a_range/beta, c_range, 'b-', lw=2)
        if alpha > beta:
            axes[0].axvline(alpha/beta, color='red', ls='--', lw=2,
                           label=f'α/β={alpha/beta:.2f}→c*={c_star:.3f}')
        axes[0].axvline(2.0, color='g', ls=':', lw=1.5, alpha=0.8, label='α=2β (canonical)')
        axes[0].axvspan(1,4,alpha=0.07,color='g')
    axes[0].set_xlabel('α/β'); axes[0].set_ylabel('c*/c_ref')
    axes[0].set_title(f'c*={c_star:.4f} | {status}')
    axes[0].legend(fontsize=8); axes[0].grid(alpha=0.3); axes[0].set_xlim(1,4)
    
    # G_N vs v0
    v0r = np.logspace(17.5,19,200)
    v0Jr = v0r*GeV_kg*c_ref**2
    G_r = hbar_ref*c_ref**3/(8*np.pi*v0Jr**2*max(alpha-beta,0.001))
    axes[1].loglog(v0r, G_r/G_ref, 'b-', lw=2)
    axes[1].axvline(v0_GeV, color='red', ls='--', lw=2,
                    label=f'G/G_ref={G_ratio:.4f}')
    axes[1].axhline(1.0, color='gray', ls=':', lw=1.5, label='G=G_ref')
    axes[1].set_xlabel('v₀ [GeV]'); axes[1].set_ylabel('G_N/G_ref')
    axes[1].set_title('G_N vs condensate VEV')
    axes[1].legend(fontsize=8); axes[1].grid(alpha=0.3)
    
    # Parameter space diagram
    b_r = np.linspace(0.1,3,100)
    axes[2].fill_between(b_r, b_r, 4*b_r, alpha=0.25, color='g', label='Lorentzian window')
    axes[2].plot(b_r, 2*b_r, 'g--', lw=2, label='α=2β (c*=c)')
    axes[2].scatter([beta],[alpha], c='red', s=120, zorder=5,
                    label=f'(β={beta:.1f}, α={alpha:.1f})')
    axes[2].set_xlabel('β'); axes[2].set_ylabel('α')
    axes[2].set_title('Lorentzian window β<α<4β')
    axes[2].legend(fontsize=8); axes[2].grid(alpha=0.3)
    axes[2].set_xlim(0,3); axes[2].set_ylim(0,9)
    plt.suptitle(f'v₀={v0_GeV:.2e} GeV | α={alpha:.2f} | β={beta:.2f}', y=1.02)
    plt.tight_layout(); plt.show()
    print(f'c*/c_ref = {c_star:.5f}  (=1 at α=2β)')
    print(f'G_N/G_ref = {G_ratio:.5f}  (=1 at v₀=M̄_Pl=2.435e18 GeV)')
    print(f'Window β<α<4β: {beta:.2f} < {alpha:.2f} < {4*beta:.2f}  →  {status}')

interact(plot_params,
  log10_v0=FloatSlider(value=18.39,min=17.5,max=19.0,step=0.01,
                       description='log₁₀(v₀/GeV):',continuous_update=False,
                       style={'description_width':'initial'}),
  alpha=FloatSlider(value=2.0,min=1.01,max=3.9,step=0.01,description='α:',continuous_update=False),
  beta=FloatSlider(value=1.0,min=0.5,max=2.0,step=0.05,description='β:',continuous_update=False));
```
