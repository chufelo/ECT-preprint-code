## Three condensate parameters → three fundamental constants

The ECT action for orientational perturbations around $\langle n_A\rangle=\delta_{Aw}$ is:

$$\mathcal{L}_E = \tfrac{1}{2}(\partial\Phi)^2 - \tfrac{\alpha}{2}(n^A\partial_A\Phi)^2 + V(n^2)$$

Perturbations see the effective metric $\mathrm{diag}(-(\alpha-1),1,1,1)$.
**Lorentzian condition:** $\alpha > 1$. Effective speed: $c_* = 1/\sqrt{\alpha-1}$.

| Condensate parameter | ECT relation | Physical constant |
|---|---|---|
| $\alpha > 1$ | $c_* = 1/\sqrt{\alpha-1}$ | Speed of light $c$ |
| $v_0 = \bar{M}_{\rm Pl}$ | $G_N = c_*^2(\alpha-1)/(16\pi v_0^2)$ | Newton $G_N$ |
| $v_0,\,\lambda$ | $\hbar_{\rm eff} = v_0^2 m_\varphi/2$ | Planck $\hbar$ |

> These are **parameter-matching relations** (Level B), not zero-parameter predictions.
> See paper §4 for the honest assessment.

```python
import numpy as np, matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider
%matplotlib inline

G_ref    = 6.674e-11   # m^3 kg^-1 s^-2
hbar_ref = 1.0546e-34  # J·s
c_ref    = 2.998e8     # m/s
GeV_kg   = 1.7827e-27  # kg per GeV/c^2
Mpl_GeV  = 2.435e18    # reduced Planck mass in GeV

def plot_params(log10_v0=np.log10(Mpl_GeV), alpha=2.0):
    v0_GeV = 10**log10_v0
    v0_kg  = v0_GeV * GeV_kg
    v0_J   = v0_kg * c_ref**2

    ok = alpha > 1.0
    if ok:
        c_star = 1.0 / np.sqrt(alpha - 1.0)
        # G_N = c*^2*(alpha-1)/(16*pi*v0^2) in natural units
        # Converting: v0 in kg*c, so v0^2 = (v0_kg*c)^2
        G_N    = c_ref**2 * (alpha - 1) / (16 * np.pi * v0_J**2 / c_ref**2)
        G_ratio = G_N / G_ref
    else:
        c_star = float('nan'); G_ratio = float('nan')

    fig, axes = plt.subplots(1, 3, figsize=(15, 4))

    # Panel 1: c* vs alpha
    a_range = np.linspace(1.01, 5.0, 400)
    c_range = 1.0 / np.sqrt(a_range - 1.0)
    axes[0].semilogy(a_range, c_range, 'b-', lw=2)
    axes[0].axvline(2.0, color='g', ls=':', lw=1.5, alpha=0.8,
                    label='α=2 → c*=1 (canonical)')
    if ok:
        axes[0].axvline(alpha, color='red', ls='--', lw=2,
                        label=f'α={alpha:.2f} → c*={c_star:.3f}')
        axes[0].scatter([alpha], [c_star], c='red', s=80, zorder=5)
    axes[0].set_xlabel('α'); axes[0].set_ylabel('c* / c_ref')
    axes[0].set_title('Effective speed  c* = 1/√(α-1)\nLorentzian: α > 1')
    axes[0].legend(fontsize=9); axes[0].grid(alpha=0.3); axes[0].set_xlim(1, 5)

    # Panel 2: G_N vs v0
    v0r     = np.logspace(17.5, 19.0, 200)
    v0Jr    = v0r * GeV_kg * c_ref**2
    G_curve = c_ref**2*(alpha-1.0)/(16*np.pi*v0Jr**2/c_ref**2) / G_ref if ok else np.nan
    axes[1].loglog(v0r, c_ref**2*max(alpha-1,0.001)/(16*np.pi*(v0r*GeV_kg*c_ref**2)**2/c_ref**2)/G_ref,
                   'b-', lw=2)
    axes[1].axvline(v0_GeV, color='red', ls='--', lw=2,
                    label=f'v₀={v0_GeV:.2e} GeV')
    axes[1].axhline(1.0, color='gray', ls=':', lw=1.5, label='G=G_ref')
    axes[1].axvline(Mpl_GeV, color='g', ls=':', lw=1.5, alpha=0.8,
                    label=f'M̄_Pl={Mpl_GeV:.2e}')
    if ok: axes[1].set_title(f'G_N vs v₀  (α={alpha:.2f})\nG/G_ref={G_ratio:.4f}')
    axes[1].set_xlabel('v₀ [GeV]'); axes[1].set_ylabel('G_N/G_ref')
    axes[1].legend(fontsize=8); axes[1].grid(alpha=0.3)

    # Panel 3: Lorentzian window in alpha space
    a_vis = np.linspace(0.5, 5, 200)
    c_vis = np.where(a_vis>1, 1/np.sqrt(np.clip(a_vis-1,1e-4,None)), np.nan)
    axes[2].plot(a_vis, c_vis, 'b-', lw=2, label='c* = 1/√(α-1)')
    axes[2].axvline(1.0, color='gray', ls='--', lw=1.5, label='α=1  (threshold)')
    axes[2].axvline(2.0, color='g',    ls=':',  lw=1.5, label='α=2  (c*=1)')
    if ok:
        axes[2].scatter([alpha], [c_star], c='red', s=120, zorder=5,
                        label=f'α={alpha:.2f}, c*={c_star:.3f}')
    axes[2].fill_betweenx([0,4], 1, 5, alpha=0.08, color='g',
                           label='Lorentzian (α>1)')
    axes[2].set_xlabel('α'); axes[2].set_ylabel('c*')
    axes[2].set_title('Lorentzian window: α > 1')
    axes[2].set_xlim(0.5, 5); axes[2].set_ylim(0, 4)
    axes[2].legend(fontsize=8); axes[2].grid(alpha=0.3)

    plt.suptitle(f'v₀={v0_GeV:.3e} GeV  |  α={alpha:.3f}  |  '
                 f'Lorentzian: {"YES" if ok else "NO (need \u03b1>1)"}', y=1.02)
    plt.tight_layout(); plt.show()
    if ok:
        print(f'c* = {c_star:.6f}  (=1 at \u03b1=2)')
        print(f'G_N/G_ref = {G_ratio:.5f}  (=1 at v\u2080=M\u0304_Pl=2.435e18 GeV)')
    else:
        print('Not Lorentzian: need \u03b1 > 1')

interact(plot_params,
  log10_v0 = FloatSlider(value=np.log10(Mpl_GeV), min=17.5, max=19.0, step=0.02,
                          description='log₁₀(v₀/GeV):',
                          continuous_update=False,
                          style={'description_width': 'initial'}),
  alpha    = FloatSlider(value=2.0, min=1.01, max=5.0, step=0.02,
                          description='α:',
                          continuous_update=False,
                          style={'description_width': 'initial'}));
```
